```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.metrics import accuracy_score
import pickle
```

```python
df=pd.read_csv('personality_dataset (1).csv')
df.isnull().sum()
```

```python
df.sample(5)
```

```python

X_train,X_test,y_train,y_test=train_test_split(df.drop('Personality',axis=1),df['Personality'],test_size=0.2,random_state=42)
```

```python
trf1=ColumnTransformer([('Simple Impute',SimpleImputer(),[0,2,3,5,6]),('Classification',SimpleImputer(strategy="most_frequent"),[1,4])])
```

```python
trf2=ColumnTransformer([('One_hot_encoding',OneHotEncoder(sparse_output=False),[1,4])],remainder='passthrough')
```

```python
trf3=ColumnTransformer([('scale',MinMaxScaler(),slice(0,10))])
```

```python
trf4= SelectKBest(score_func=chi2,k=5)
```

```python
trf5=DecisionTreeClassifier()
```

```python
pipe = Pipeline([('trf1',trf1),('trf2',trf2),('trf3',trf3),('trf4',trf4),('trf5',trf5)])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
